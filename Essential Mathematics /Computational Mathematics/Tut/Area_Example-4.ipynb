# Finding Area by Monte-Carlo Method in Julia 


$\hspace{1cm}$

## Example: Find the area of $\frac{x^2}{9}+\frac{y^2}{4}<=1$

### Graph of the Ellipse

```julia
using PyPlot
t=0:0.01*pi:2*pi
x=3*cos.(t)
y=2*sin.(t)
plot(x,y)
plot([-3,3],[0,0])
plot([0,0],[-2,2])
plot([0, 3],[2, 2],color="red")
plot([3, 3],[0, 2],color="red")
plot([0,3],[0,0],color="red")
plot([0,0],[0,2],color="red")
```

### Method Explained: 

1. (area of the ellipse) = 4$\times$(area of top-right corner of the ellipse)

2. Create $N$ random points in box $[0, 3]\times [0, 2]$ (red box in figure). Suppose there are $innn$ of the $N$ random points fall in the top-right corner of the ellipse. 

3. When $N$ is large, $\frac{area\ of\ topright\ corner\ of\ ellipse}{area\ of\ box\ [0, 3]\times [0, 2]}\approx \frac{innn}{N}$ $\ \ \Longrightarrow\ \ ${area\ of\ topright\ corner\ of\ ellipse}\approx \frac{innn}{N}\times 6$.

3. $\Longrightarrow\ \ $area\ of\ ellipse $=4\times${area\ of\ topright\ corner\ of\ ellipse}\approx 4\times\frac{innn}{N}\times 6=24\times \frac{innn}{N}$

NOTE: area of the box $[0, 3]\times [0, 2]\ =\ 6$.   

### Code

```julia
innn=0.0
outt=0.0
N=10000000
for i in 1:N
    x=rand(2,1)
    if x[1,1]^2+x[2,1]^2<=1
        innn=innn+1
    else
        outt=outt+1
    end
end
area=(innn/N)*24
println("Area=",area)
```
