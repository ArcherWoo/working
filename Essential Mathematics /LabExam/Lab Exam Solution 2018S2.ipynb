##### Q3b(i)

```julia
# Critical points of f(x) are roots of f'(x)=0, so we f'(x)=0.
# To use Newton's method, we need functions f'(x) and f''(x).
# Define functions f'(x) and f''(x).
function df(x)
    3*x.^2-0.21*exp(x)+0.5*cos(x)-12.4;
end
function ddf(x)
    6*x-0.21*exp(x)-0.5*sin(x);
end
```

```julia
# solve f'(x)=0 starting from x0=1.0
x0=3
x=x0
for i in 1:10
    x=x-df(x)/ddf(x)
end
x
```

```julia
df(x) # checking if x is realy a root for f'(x)=0
```

### Q4

```julia
using DifferentialEquations
```

```julia
using Plots
```

```julia
function f(x,p,t)  
    0.1*t.^2-t.*sin(t.-sqrt(t+1))  #input function expression in t
end
tspan=(0.0,5.0)         #timespan of the solution (input floating numbers for the time span such as (0.0, 2.0))
x0=    -2.0             #Initial values of the states 
prob=ODEProblem(f,x0,tspan);
```

```julia
soln_x=solve(prob,Tsit5(),reltol=1e-8,abstol=1e-8);
```

```julia
plot(soln_x)
```

```julia
soln_x(5)
```
