#### Q3 (a)

```julia
function ff(x)
    x.^3/3-2*x.*sin(x)+3*x-10
end
function dffdx(x)
    x.^2-2*x.*cos(x)-2*sin(x)+3
end
x0=-1.0
x=x0
for i in 1:10
    x=x-ff(x)/dffdx(x)
end
x
```

```julia
ff(x)
```

#### Q3 (b)

```julia
using DifferentialEquations
```

```julia
using Plots
```

```julia
function f(x,p,t)  
    2*t.*sin(t^2-1)
end
tspan=(0.0,5.0)                    #timespan of the solution
x0=1.0                             #Initial values of the states 
Q3_b=ODEProblem(f,x0,tspan);
```

```julia
soln_x=solve(Q3_b,Tsit5(),reltol=1e-8,abstol=1e-8)
plot(soln_x)
```

```julia
soln_x(5)
```

#### Q4 (a)

```julia
# define f'(x) and f''(x)
function df(x)
    2*x-x.*sin(x)+cos(x);
end
function ddf(x)
    2-x.*cos(x)-2*sin(x);
end
```

```julia
# solve f'(x)=0 starting from x0=1.0
x0=-2.0
x=x0
for i in 1:10
    x=x-df(x)/ddf(x)
end
x
```

```julia
df(x) # checking if x is realy a root for f'(x)=0
```

```julia
x10=x; # x10=-0.38892154717566935 is an approximate critical point.
```
