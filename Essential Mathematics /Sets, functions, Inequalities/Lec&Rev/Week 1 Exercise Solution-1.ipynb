Find the union of sets $ùê¥$ and $ùêµ$ where $ùê¥=\{1,2,3,4\}$ and $ùêµ=\{4,5,6,7\}$

```julia
A=1:4    #the array of numbers from 1, 
         #incraesing by 1 (defaut) each time, 
         #till reaching 4
         #If increasing by 0.1 each time, the code would be
         #A=1:0.1:4
B=4:7
union(A,B)
```

### Q1. 
Locate the zeros of the polynomial function x^3-6x^2+11x-6 by plotting the function.

```julia
using PyPlot
x=0:0.01:4
y=x.^3-6*x.^2+11*x-6
plot(x,y)
grid()
```

Zeros: x=1, x=2, x=3

### Q3. 
Given two polynomial functions $ùë¶_1=ùë•^4‚àí6ùë•^3+11ùë•^2‚àí6ùë•$ and $ùë¶_2=ùë•^2‚àí3ùë•‚àí4$. Plot these functions on the interval $[0,3]$. Do either of the polynomials have any roots in this interval?

```julia
x=0:0.01:3
y1=x.^4-6x.^3+11x.^2-6x
y2=x.^2-3x-4
plot(x,y1,label="y1")
plot(x,y2,label="y2")
grid()
legend()
```

Zeros for $y_1$: 0, 1, 2, 3; no zero for $y_2$

### Q4.
The rational function $z=\frac{x^4-6x^3+11x^2-6x}{x^2-3x-4}$ is defined over the domain $[0,3]$. Plot this function. Why are the roots the same as $y_1$ above?


```julia
x=0:0.01:3
z=(x.^4-6*x.^3+11*x.^2-6*x)./(x.^2-3*x-4)
plot(x,z)
grid()
```

Zero occurs at $x=0$

### Q5.

Plot in the same graph with the power function $ùë•^3$ and the power function 3ùë• over the interval $[0,4]$. Which grows faster?

```julia
x=0:0.01:4
y1=x.^3
y2=3.^x
plot(x,y1,label="y=x^3")
plot(x,y2,label="y=3^x")
grid()
legend()
```

It is clear that $3^x$ grows faster.
