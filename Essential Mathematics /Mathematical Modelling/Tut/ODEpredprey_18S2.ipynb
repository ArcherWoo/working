```julia
# We need two packages for this problem
using DifferentialEquations     
using Plots 
```

```julia
# The system of ODEs to be solved: 
function f(dudt,u,p,t) 
    r=0.1; b=0.01; c=0.001; d=0.1;   #These are parameters used in the model 
    dudt[1]=r*u[1]-b*u[1]*u[2]       # preditor population
    dudt[2]=c*u[1]*u[2]-d*u[2]       # prey polulation
end
tspan=(0.0,200.0)                    #timespan of the solution
u0=[50.0;20.0]                       #Initial values of the states 
predprey=ODEProblem(f,u0,tspan);
```

```julia
res = solve(predprey,reltol=1.0e-8); # Solves the problem using 
                                             # Tsitouras method 
```

```julia
# The results can now be plotted
plot(res,xlabel="Time",ylabel="Population",title="Predator(u1)-prey(u2) system")
#legend(["prey","rredator"]) 
```

```julia
plot(res,vars=(1,2),xlabel="Prey",ylabel="Predator",title="Predator Prey phase plane")
```
