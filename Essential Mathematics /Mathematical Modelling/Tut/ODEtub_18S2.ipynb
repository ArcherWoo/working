```julia
# We need two packages for this problem
using DifferentialEquations    # This contains the algorithm to solve numerically the differential equations
using Plots  
```

```julia
# Define the differential equation
f(u,p,t)=3.0;                # This is what the function that returns the rate of change 
tspan=(0.0,5.0)                 # specify the timespan over which the solution is required
u0=0.0                          # specify the initial value of u (ie empty bathtub)
cisternP=ODEProblem(f,u0,tspan) # the ode problem
volume = solve(cisternP);       # Solves the problem returning  
```

```julia
plot(volume,xlabel="time",ylabel="volume",title="Volume of water in bathtub")
```
