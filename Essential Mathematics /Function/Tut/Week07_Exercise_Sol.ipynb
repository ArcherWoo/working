## Exercise 7.1.2 
### Q3(a)

```julia
# Method 1. Plot x/sin(x) from -1 to +1
x=-1:0.05:1
y=x./sin.(x)    #Note the 'dot'
plot(x,y)
xlabel("x")
ylabel("y")
title("y=x/sin(x)")
```

####  Note from the figure that although x/sin(x) is not defined at x=0
#### It approachs the value 1.00 whether approaching 0.0 from the left (through negative x-values) or from the right (through positive x-values)

```julia
# Method 2. Print out values.
# First print out a sequence of x values approaching 0.0 from the right and their corresponding y=x/sin(x) values
for x=0.50:-0.05:0.05
    println("x=",x, "    ","y=",x./sin(x)) 
end
```

#### The results indicate the values approaching 1.0 as x approaches 0 from the right.

```julia
# Then print out a sequence of x values approaching 0.0 from the left and their corresponding y=x./sin(x) values
for x=-0.5:0.05:-0.05
    println("x=",round(x,2), "    ","y=",sin(x)./x)
end
```

#### The results indicate the values approaching 1.0 as x approaches 0 from the left. 
#### In summary, x/sin(x) tends to 1 as x approaches to 0.

## Exercise 7.1.2
### Q3(b).

```julia
#Method 1. Plotting 
x=-1.0:0.01:1.0
y=(sin.(x)).^2./(x.^2)
plot(x,y)
xlabel("x")
ylabel("y")
title("y=sin^2(x)/x^2")
```

Note that although $sin^2(x)/x^2$ is not defined at $x=0$, it approachs the value 1.00 when $x \to 0.0$.

So, the limit sought is 1

```julia
# Method 2. Printing out values.
x=[-0.1:0.01:-0.01;0.01:0.01:0.1]
y=(sin.(x)).^2./(x.^2)
for i=1:length(x)
    println(x[i],"       ",y[i])
end
```

####It can be seen that when x approaches to 0, sin^2(x)/x^2 tends to 1

## Exercise 7.1.4 
### Question 5

```julia
# We want to see what happens when x becomes very large
# Let us see what happens when x goes from 1 million to 5 million
x=1.0e6:1.0e7:1e8
y=(1+1./x).^x
for i=1:length(x)
    println(x[i],"      ",y[i])
end
```

### This seems to be converging to a finite number. Look familiar? See below.

```julia
e
```

####Thus, the limit is e
####To see this another way, you can do the following

```julia
x=1.0e6:1.0e7:1e8
y=(1+1./x).^x-e
for i=1:length(x)
    println(x[i],"      ",y[i])
end
```
